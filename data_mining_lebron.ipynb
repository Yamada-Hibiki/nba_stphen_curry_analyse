```python
import pandas as pd
import glob
import os
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set_style('whitegrid')
```

```python
folder_path = "data//lebron_data"

# Use glob to get a list of all Excel files in the folder
csv_files = glob.glob(f"{folder_path}/*.csv")

# Loop through each CSV file and read it into a Pandas DataFrame
dfs = []
for file in csv_files:
    df = pd.read_csv(file, sep=";")
    print(file, df.shape)
    year = df["Date"].iloc[0].split("/")[2]
    df["Season"] = year + "-" + str(int(year) + 1)
    dfs.append(df)

# If you want to concatenate all DataFrames into a single DataFrame, you can use pd.concat
all_games_df = pd.concat(dfs, ignore_index=False)

all_games_df.to_csv("data//game_logs_lebron", index=False)
```

```python
def return_rk_number_counter():
    return rk_number_counter
```

```python
def replace_none_to_number(number):
    global rk_number_counter
    if not np.isnan(number):
        return number
    else:
        rk_number_counter = return_rk_number_counter()
        rk_number_counter += 1
        return rk_number_counter
```
