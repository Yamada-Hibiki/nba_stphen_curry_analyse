```python
import pandas as pd
import glob
import os
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set_style('whitegrid')
```

```python
folder_path = "data//steph_data_csv"

# Use glob to get a list of all Excel files in the folder
csv_files = glob.glob(f"{folder_path}/*.csv")

# Loop through each CSV file and read it into a Pandas DataFrame
dfs = []
for file in csv_files:
    df = pd.read_csv(file, sep=";")
    print(file, df.shape)
    dfs.append(df)

# If you want to concatenate all DataFrames into a single DataFrame, you can use pd.concat
all_games_df = pd.concat(dfs, ignore_index=False)

all_games_df.to_csv("data//game_logs_steph", index=False)
```

```python
player_df = pd.read_csv("data//game_logs_steph", sep=",")
player_df.head()
```

```python
def get_season(date):
    month = date.split("/")[1]
    year = date.split("/")[2]

    if 1 <= int(month) <= 5:
        year = str(int(year) - 1)

    season = year + "-" + str(int(year) + 1)
    return season
```

```python
player_df["Season"] = player_df["Date"].apply(get_season)
player_df.head()
```

```python
player_df["Year"] = player_df["Season"].apply(lambda x: x[:4]).astype(int)
player_df.head()
```

```python
player_df["Rk"] = player_df["Rk"].astype(str)
```

```python
player_df["Season-game-number"] = player_df["Season"] + "-" + player_df["Rk"]
```

```python
team_df = pd.read_csv("data//stphen_team_logs", sep=",")
team_df.head()
```

```python
month_list = ["January", "February", "March", "April", "May", "June", "July", "August", 
              "September", "October", "November", "December"]

month_dic = {}
for index, month in enumerate(month_list):
    month_dic[month[:3]] = str(index + 1)
print(month_dic)
```

```python
def get_season_gsw(date):
    year = date.split(", ")[2]
    print(year)
    date_string = date.split(", ")[1].split(" ")[0]
    print(date_string)
    month = month_dic[date_string]

    if 1 <= int(month) <= 5:
        year = str(int(year) - 1)

    season = year + "-" + str(int(year) + 1)
    return season
```

```python
print(get_season_gsw("Tue, Oct 18, 2022"))
```

```python
team_df["Season"] = team_df["Date"].apply(get_season_gsw)
```
